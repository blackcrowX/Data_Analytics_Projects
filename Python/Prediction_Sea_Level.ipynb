<h1 align="center">Introduction</h1>

We will analyze a dataset of the global average sea level change since 1880. We will use the data to predict the sea level change through year 2050. The dataset was published by <a href="https://www.freecodecamp.org/learn/data-analysis-with-python/data-analysis-with-python-projects/sea-level-predictor">freeCodeCamp</a>.

<h1 align="center">Setup</h1>

## Step 1: Import Libraries

Import and configure libraries required for data analysis.

```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import linregress
```

## Step 2: Import Dataset

Use Pandas to import the data from `epa-sea-level.csv`.

```python
url = "https://raw.githubusercontent.com/blackcrowX/Data-Analysis-Projects/main/Python/Sea_Level_Predictor/epa_sea_level.csv"
df = pd.read_csv(url)
```

<h1 align="center">Data Visualisation</h1>

## Step 3: Create Scatter Plot

Use matplotlib to create a scatter plot using the `Year` column as the x-axis and the `CSIRO Adjusted Sea Level` column as the y-axis.

```python
f, ax = plt.subplots(1,1)
ax = plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
```

## Step 4: Create First Line Best Fit

Use the `linregress` function from `scipy.stats` to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
f, ax = plt.subplots(1,1)
lineA = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
xA = np.arange(df['Year'].min(),2050,1)
yA = xA*lineA.slope + lineA.intercept
ax = plt.plot(xA,yA)
```

## Step 5: Create Second Line Best Fit

Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.

```python
f, ax = plt.subplots(1,1)
df_2000 = df[df['Year'] >= 2000]
lineB = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])
xB = np.arange(2000,2050,1)
yB = xB*lineB.slope + lineB.intercept

ax = plt.plot(xB,yB)
```

## Step 6: Add Labels & Save Plot

The x label should be `Year`, the y label should be `Sea Level (inches)`, and the title should be `Rise in Sea Level`. Then save the plot.

```python
ax = plt.plot(xB,yB)
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.savefig("Sea_Level_Predictor")
```
